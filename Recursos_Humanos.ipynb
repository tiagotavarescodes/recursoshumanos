Para utilizar os modelos de aprendizado de m√°quina √© necess√°ria converter os dados categ√≥ricos que est√£o no formato string para num√©rico.

```python
X_cat = df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotenconder = OneHotEncoder()
X_cat = onehotenconder.fit_transform(X_cat).toarray()
```

```python
type(X_cat)
```

```python
X_cat = pd.DataFrame(X_cat)
type(X_cat)
```

```python
X_cat.shape
```

```python
X_cat
```

```python
X_numerical = df[['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating', 'RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
```

```python
X_all.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_all.columns = X_all.columns.astype(str)
X = scaler.fit_transform(X_all)
```

```python
X
```

```python
y = df['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape[0]
```

```python
X_test.shape[0]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='macro')
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
y_pred = forest.predict(X_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
```
